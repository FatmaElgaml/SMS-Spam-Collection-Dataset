### This helps us automatically identify the correct encoding form

### Using the correct encoding ensures that the data is read properly (avoid Unicode errors or corrupted text)

### This step improves data quality and helps achieve better model performance.

```python
from webencodings import Encoding

import chardet
import pandas as pd

with open("spam.csv","rb") as data:
    kind_of_encoding = chardet.detect(data.read())['encoding']



df = pd.read_csv("spam.csv", encoding= kind_of_encoding)
print(kind_of_encoding)
```

### Remove unnecessary "Unnamed" columns created due to empty columns in CSV.
### These columns contain no useful information for the model.

## to drop all columns prefix "Unnamed"         -->  df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.head(15)
```

```python
df.columns = ['label', 'message']
df.head(15)
```

```python
import re

def clean_text(text):

    text = text.lower()                             # all letter small

    text = re.sub(r'http\S+', '', text)             #remove links 

    text = re.sub(r'\d+', '', text)                 #remove numbers 

    text = re.sub(r"[^\w\s']", '', text)             #remove not letter, numbers and spase and '  (\w\s = \W)

    text = re.sub(r'\s+', ' ', text).strip()        #remove Extra Space 

    return text
```

```python
df["clean_message"] = df["message"].apply(clean_text)
df.head(15)
```

### Now we do Tokenization 

```python
! pip install nltk
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
df["word_token"] = df["clean_message"].apply(word_tokenize)
df.head(15)
```

### part of stop word removal

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
print(stop_words)

```

```python
df["word_token"] = df["word_token"].apply(
    lambda words: [word for word in words if word not in stop_words])
df.head(15)
```
